```python
# StringDataType - nhập 1 chữ cái thường, in mã Unicode và chữ hoa tương ứng
ch = input().strip()
if len(ch) != 1:
    print("Vui lòng nhập đúng 1 kí tự")
else:
    # in mã Unicode (code point)
    print(ord(ch))
    # chuyển thành chữ hoa và in
    print(ch.upper())
```
