```python
# Bubble Sort
def bubble_sort(arr):
    a = arr[:]
    for i in range(len(a)):
        for j in range(len(a)-1-i):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a

# Selection Sort
def selection_sort(arr):
    a = arr[:]
    for i in range(len(a)):
        m = i
        for j in range(i+1, len(a)):
            if a[j] < a[m]:
                m = j
        a[i], a[m] = a[m], a[i]
    return a
```

```python
def most_frequent(arr):
    count = {}
    for x in arr:
        count[x] = count.get(x, 0) + 1
    max_freq = max(count.values())
    for x in arr:
        if count[x] == max_freq:
            return x, max_freq
```

```python
def count_pairs(arr, X):
    cnt = 0
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] + arr[j] == X:
                cnt += 1
    return cnt
```

```python
def length_of_LIS(nums):
    dp = [1]*len(nums)
    for i in range(len(nums)):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j]+1)
    return max(dp) if nums else 0
```

```python
def min_interval(intervals, queries):
    res = []
    for q in queries:
        best = float('inf')
        for l, r in intervals:
            if l <= q <= r:
                best = min(best, r-l+1)
        res.append(best if best != float('inf') else -1)
    return res
```

```python
def count_smaller(nums):
    res = []
    for i in range(len(nums)):
        cnt = 0
        for j in range(i+1, len(nums)):
            if nums[j] < nums[i]:
                cnt += 1
        res.append(cnt)
    return res
```

```python
def max_subarray_mod(nums, m):
    pref = 0
    best = 0
    seen = [0]
    for x in nums:
        pref = (pref + x) % m
        best = max(best, pref)
        for s in seen:
            best = max(best, (pref - s) % m)
        seen.append(pref)
    return best
```
